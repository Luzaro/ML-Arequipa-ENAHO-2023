# Modelado Baseline

Notebook para reproducir y revisar el modelado baseline antes del tuning.

```python
import sys
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
SRC_PATH = PROJECT_ROOT / 'src'
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))
```

```python
from utils.paths import build_paths
from modeling.baseline import load_inputs, detect_feature_sets, run_feature_set

paths = build_paths(PROJECT_ROOT)
dataset, top20_df = load_inputs(paths)
top30_full, _, missing = detect_feature_sets(dataset, top20_df)

print('Dataset:', dataset.shape)
print('Top30 full:', len(top30_full))
print('Missing:', missing)
```

```python
summary, details_full, meta_full = run_feature_set(dataset, 'top30_full', top30_full)

summary = (
    summary
    .sort_values(['recall_pobre', 'f1_pobre', 'accuracy'], ascending=[False, False, False])
    .reset_index(drop=True)
)
summary
```

```python
best_recall = summary.iloc[0]
best_balanced = summary.sort_values(['f1_pobre', 'recall_pobre', 'precision_pobre'], ascending=[False, False, False]).iloc[0]

print('Mejor baseline por recall:')
print(best_recall)
print('\nMejor baseline balanceado:')
print(best_balanced)
```
