```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
DATA_PATH = PROJECT_ROOT / 'data' / 'processed' / 'enaho_arequipa_escenario_b_clean.csv'
REPORTS_EDA = PROJECT_ROOT / 'reports' / 'eda'

df = pd.read_csv(DATA_PATH)
df.shape
```

```python
target_counts = df['target_pobreza_monetaria_bin'].value_counts(dropna=False).sort_index()
target_counts
```

```python
sns.set_theme(style='whitegrid')
plt.figure(figsize=(6, 4))
sns.barplot(x=target_counts.index.astype(str), y=target_counts.values)
plt.title('Distribución del target')
plt.xlabel('target_pobreza_monetaria_bin')
plt.ylabel('Número de hogares')
plt.show()
```
